```python
# Importing libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import sklearn
```

```python
df = pd.read_csv('amazon.csv')
print(df.head())
```

```python
df['category'].unique() # 211 categories



df['category_list'] = df['category'].astype(str).str.split('|')
cat_df = df.explode('category_list')
cat_df['category_list'].nunique() #317 total categories

# Plot number of products in each overall_category
top_cats = cat_df['category_list'].value_counts().head(20)

plt.figure(figsize=(10,6))
sns.barplot(x=top_cats.values, y=top_cats.index)
plt.title("Top 20 Product Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

```python
#clean rating count
df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
df['rating_count'].head()
```

```python
# Clean Prices (convert to USD and numeric)
def clean_price(price):
  if not isinstance(price, str):
    return 0.0
  price = price.replace('â‚¹', '')
  if ',' in price:
    price = price.replace(',', '')
  ans = float(price)
  return ans * 0.011

df['actual_price'] = df['actual_price'].apply(clean_price)
df['discounted_price'] = df['discounted_price'].apply(clean_price)

df[['actual_price', 'discounted_price']].head(30)
```

```python
# Distribution of ratings

df['clean_rating'] = df['rating'].str.extract(r'(\d+\.\d+)')
df['clean_rating'] = pd.to_numeric(df['clean_rating'], errors='coerce')

# Plot distribution
plt.figure(figsize=(10,6))
sns.histplot(df['clean_rating'], bins=20, kde=True)
plt.title("Distribution of Ratings")
```

```python
# Correlation heatmap
corr = df[['rating_count', 'discounted_price', 'actual_price', 'clean_rating']].corr() #add more features soon

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Numeric Features Heatmap')
plt.show()
```
